```python
random_numbers = [15, 81, 5, 17, 20, 21, 13]
count = 0
divisable_by_five = []

for number in random_numbers:

    if number % 5 == 0:
        divisable_by_five.append(number)
    else:
        continue

print("all numbers divisible by 5 in reverse order:" )

divisable_by_five.sort(reverse=True)
for number in divisable_by_five:
    count += 1
    print(f"{count}. {number}")
```

```python
my_ranks = {
    'Math': 'A',
    "Science": 'B',
    'Drawing': 'A',
    'Sports': 'C'
    }

rank_points = {
    "A": 100,
    "B": 80,
    "C": 40
}

total_points = 0

for subject, rank in my_ranks.items():
    print(f"My Rank in {subject} => {rank} => {rank_points[rank]} Points")
    points = rank_points[rank]
    total_points += points
print(f"Total Points => {total_points}")
```
